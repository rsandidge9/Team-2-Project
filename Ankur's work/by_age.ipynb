```python
# Dependencies and Setup
import pandas as pd
import scipy.stats as st

# Study data files
by_age_path = "data/by_age_data.csv"

# Read the mouse data and the study results
by_age_results = pd.read_csv(by_age_path)
by_age_results
```

```python
year_by_year = pd.DataFrame(by_age_results)
year_by_year
```

```python
import matplotlib.pyplot as plt
import numpy as np

#year_by_year.set_index('Year').plot(figsize=(12,12));



#year_by_year.plot(kind='line', x="Year", y = {"16-19 years","20-24 years", "25-34 years", "35-44 years",
                                             # "45-54 years", "55-64 years", "65 years and older"} ,
                  #title ='Occupational diversity', figsize=(18, 18))

#plt.xticks(np.arange(1979,2013,1))

x = year_by_year['Year']

y1 = year_by_year['Total, 16 years and older']
y2 = year_by_year['16-19 years']
y3 = year_by_year['20-24 years']
y4 = year_by_year['25-34 years']
y5 = year_by_year['35-44 years']
y6 = year_by_year['45-54 years']
y7 = year_by_year['55-64 years']
y8 = year_by_year['65 years and older']


linestyle1 = '--'
linestyle2 = '-.'
linestyle3 = ':'

plt.figure(figsize=(25,25))

plt.plot(x,y1, color = 'blue')
plt.text(1979, 62, 'Age:Total', fontsize=15, color = 'blue')

plt.plot(x, y2, color = 'green',linestyle=linestyle1)
plt.text(1985, 94, 'Age:16-19 years',fontsize=15,color = 'green')

plt.plot(x, y3, color = 'red')
plt.text(1981, 81, 'Age:20-24 years', fontsize = 15, horizontalalignment='right',color = 'red')

plt.plot(x, y4, color = 'orange')
plt.text(1988, 78, 'Age:25-34 years', fontsize = 15, horizontalalignment='right', color = 'orange')

plt.plot(x, y5, color = 'grey')
plt.text(1986, 63, 'Age:35-44 years', fontsize = 15, horizontalalignment='right',color = 'grey')

plt.plot(x, y6, color = 'black', linestyle = linestyle2)
plt.text(1992, 68, 'Age:45-54 years', color = 'black',fontsize = 15)

plt.plot(x, y7, color = 'magenta')
plt.text(1997, 64, 'Age:55-64 years', color = 'magenta', fontsize = 15)

plt.plot(x, y8, color = 'brown',linestyle =linestyle3)
plt.text(1982, 74, 'Age:65 and above', color = 'brown', fontsize = 15,horizontalalignment='right')



plt.xticks(np.arange(1979,2013,1))

plt.xlabel("Years", fontsize = 20)
plt.ylabel("Percent", fontsize = 20)
plt.title('Women earning as percent of men for last 30 years', fontsize = 30)

plt.savefig("by_age.png")
```
