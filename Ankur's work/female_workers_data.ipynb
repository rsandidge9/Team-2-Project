```python
# Dependencies and Setup
import pandas as pd
import scipy.stats as st

# Study data files
Female_path = "data/Female_workers.csv"

# Read the mouse data and the study results
Female_results = pd.read_csv(Female_path)
Female_results
```

```python
female_workers = pd.DataFrame(Female_results)
female_workers_df = female_workers.rename(columns={"Number\nof\nworkers": "Total Number of Females",
                                                 "Median\nweekly\nearnings": "Median Earnings(total)"}).drop(0)

#men_workers_df.dtypes

female_numbers = female_workers_df["Total Number of Females"].str.replace(",","").astype(int)

new_female_df = female_workers_df.drop(columns = ['Total Number of Females'])
new_female_df

female_numbers_df = pd.DataFrame(female_numbers)
female_numbers_df

final_female_df = pd.concat([new_female_df, female_numbers_df], 1)
final_female_df

sorted_female_df = final_female_df.sort_values(by=["Total Number of Females"], ascending = False)

female_df = sorted_female_df[['Occupation','Total Number of Females','Median Earnings(total)', 'Standard error of median' ]]
female_df
```

```python
import matplotlib.pyplot as plt

female_df.plot(kind='bar', x="Occupation", y="Total Number of Females", 
              color = 'green', title ='Occupational diversity', alpha=0.5, align="center", figsize=(18,5))

plt.xlabel("Occupation")
plt.ylabel("Total Number of Females")
plt.show()

plt.savefig("Occupational Diversity_Females")
```
