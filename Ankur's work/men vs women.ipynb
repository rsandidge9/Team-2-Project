```python
# Dependencies and Setup
import pandas as pd
import scipy.stats as st
import matplotlib.pyplot as plt
import numpy as np

# Study data files
Men_vs_women_path = "data/Men vs women.csv"

# Read the mouse data and the study results
Men_vs_women = pd.read_csv(Men_vs_women_path)
Men_vs_women
```

```python
Men_vs_women_raw = pd.DataFrame(Men_vs_women)

Men_vs_women_df = Men_vs_women_raw.dropna()

worker_numbers = Men_vs_women_df["Number of workers"].str.replace(",","").astype(int)

workers_df = pd.DataFrame(worker_numbers)
workers_df

earning_numbers = Men_vs_women_df["Median weekly earnings"].str.replace("$","").astype(int)
earning_df = pd.DataFrame(earning_numbers)

gender_and_year_format = Men_vs_women_df["Characteristic"].rename(" Year and Gender ")
gender_and_year = pd.DataFrame(gender_and_year_format)

final_df = pd.concat([gender_and_year, workers_df, earning_df], 1)
final_df
```

```python


df_plot = pd.DataFrame({'Gender': ['2018 Female', '2018 Male'],
                   'Number of workers': [51425, 64142]},  index=['Female', 'Male'])

color = ["blue", "magenta"]

plot = df_plot.plot.pie(y='Number of workers', figsize=(5, 5), colors=color, autopct='%1.1f%%', startangle=60, explode=(0.04,0.03))

plt.title("2018: Working Male va Female numbers")

plt.text(1,2, "Total number of workers in 2018 were 115,567",  fontsize = 15, horizontalalignment='right')
plt.show()
```
