```python
# Dependencies and Setup
import pandas as pd
import scipy.stats as st

# Study data files
men_path = "data/Men_workers.csv"

# Read the mouse data and the study results
men_results = pd.read_csv(men_path)
men_results
```

```python
men_workers = pd.DataFrame(men_results)
men_workers_df = men_workers.rename(columns={"Number\nof\nworkers": "Total Number of Men",
                                                 "Median\nweekly\nearnings": "Median Earnings(total)"}).drop(0)

#men_workers_df.dtypes

men_numbers = men_workers_df["Total Number of Men"].str.replace(",","").astype(int)

new_men_df = men_workers_df.drop(columns = ['Total Number of Men'])
new_men_df

men_numbers_df = pd.DataFrame(men_numbers)
men_numbers_df

final_men_df = pd.concat([new_men_df, men_numbers_df], 1)
final_men_df

sorted_df = final_men_df.sort_values(by=["Total Number of Men"], ascending = False)

men_df = sorted_df[['Occupation','Total Number of Men','Median Earnings(total)', 'Standard error of median' ]]
men_df
```

```python
import matplotlib.pyplot as plt

final_df.plot(kind='bar', x="Occupation", y="Total Number of Men", 
              color = 'blue', title ='Occupational diversity', alpha=0.5, align="center", figsize=(18,5))

plt.xlabel("Occupation")
plt.ylabel("Total Number of Men")
plt.show()
```
