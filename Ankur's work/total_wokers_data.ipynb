```python
# Dependencies and Setup
import pandas as pd
import scipy.stats as st

# Study data files
total_results_path = "data/Total_workers.csv"

# Read the mouse data and the study results
total_results = pd.read_csv(total_results_path)
total_results
```

```python
total_workers = pd.DataFrame(total_results)
total_workers_df = total_workers.rename(columns={"Number\nof\nworkers": "Total Number of Workers",
                                                 "Median\nweekly\nearnings": "Median Earnings(total)"}).drop(0)

#total_workers_df.dtypes

#total_workers_df.sort_values(by=["Total Number of Workers"], ascending = True)

worker_numbers = total_workers_df["Total Number of Workers"].str.replace(",","").astype(int)

new_workers_df = total_workers_df.drop(columns = ['Total Number of Workers'])
new_workers_df

workers_df = pd.DataFrame(worker_numbers)
workers_df

final_df = pd.concat([new_workers_df, workers_df], 1)
final_df

final_df.sort_values(by=["Total Number of Workers"], ascending = False)

final_df = final_df[['Occupation','Total Number of Workers','Median Earnings(total)', 'Standard error of median' ]]
final_df
```

```python
import matplotlib.pyplot as plt

#x_values = final_df['Occupation']
#y_values = final_df['Total Number of Workers']


final_df.plot(kind='bar', x="Occupation", y="Total Number of Workers", 
              color = 'blue', title ='Occupational diversity', alpha=0.5, align="center", figsize=(18,5))

plt.xlabel("Occupation")
plt.ylabel("Total Number of Workers")
plt.show()
```

```python
pie_chart = final_df.plot.pie( y ="Total Number of Workers", labels=final_df["Occupation"],autopct="%1.1f%%",
                                      title ="Gender Ratio", figsize=(28, 50))
```
